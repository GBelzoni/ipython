```python
#Create Binary features, e.g. has a political keyword
####Not currently using in classification
#Clean text
text2 = text.str.replace('\r|\n', ' ')
# text_comments2[text_comments2.str.contains( '\n')]

has_poli = text2.str.contains( '[Rr][Aa][Oo]|[Rr]aila|[Uu]huru|[Rr]uto|[Uh]uruto|[\.\s][Cc]ord(ed)*|[Jj]ubile+')
#print "has poli \n", has_poli.value_counts()
has_tribal = text2.str.contains( '[Ll][Uu][Oo]|[Kk][Ii][Kk]([Uu][Yy][Uu])|[Kk][Ii][Uu[Kk]|[Kk]alenj|[Tribe]')
# print "has tribal \n", has_tribal.value_counts()
has_exclamation = text2.str.contains( '\!')
# print "has exclamation \n", has_exclamation.value_counts()
has_danger_word = text2.str.lower().str.contains( 'thief|mwezi|greed|hate|thiev|sham|jigg|devil|bastard|fuck|penis|rig')
# print "has danger \n", has_danger_word.value_counts()
pd.DataFrame(text2[has_danger_word])

#Binary features
df_binary_features = pd.DataFrame(has_poli)
df_binary_features.columns = ['HAS_POLI']
df_binary_features['HAS_TRIBAL']= has_tribal
df_binary_features['HAS_EXCL']= has_exclamation
df_binary_features['HAS_DANGERW']= has_danger_word
df_binary_features.head()
#df_binary_features = pd.DataFrame([has_poli,has_tribal,has_exclamation,has_danger_word],columns=['HAS_POLI','HAS_TRIBAL','HAS_EXCL', 'HAS_DANGERW'])
```

```python
#Create text/word vectorised features

#Input variables
UseTfidf = True #Set to false to use count vect
Tfidf_max_df = 0.95
ngram_range = (1,2)
stop_words = 'english'

from sklearn.feature_extraction.text import CountVectorizer #Unormalised word frequencies
from sklearn.feature_extraction.text import TfidfVectorizer

###Feature Comments Vecotoriser
#This extract text at numpy array
text_comments = text.values

#There are many possible args to Vectorisers - tab inside first parantheses to get list

if UseTfidf:
    vectoriser_comments = TfidfVectorizer(stop_words='english',
                                          strip_accents='unicode',                    
                                          ngram_range = ngram_range,
                                          max_df = Tfidf_max_df
                    )
else:
    vectoriser_comments = CountVectorizer(stop_words='english',
                                          strip_accents='unicode',
                                          ngram_range = ngram_range)

t = time.time()
features_vect_text = vectoriser_comments.fit_transform(text_comments) 
print "training comments to word vector took", time.time()-t, "seconds"
print "features comments dimension ", features_vect_text.shape
#Change from sparse matrix to dense matrix
#vect_train = features_msg_training.todense() #BIG memory usage 2GB, see how to use the sparse for training
```
